```python
from matplotlib import pyplot as plt
import numpy as np, pandas as pd, seaborn as sns
import os, json
```

```python
with open('../../config.json') as file:
    DNAME_OUT = json.load(file)['outdir']

read_csv = lambda x, **k: pd.read_csv(os.path.join(DNAME_OUT, x + '_final.csv'), **k)
runs = read_csv('runs')
runs = runs.set_index('run')
```

```python
responses = read_csv('responses', on_bad_lines='skip')
for i, row in responses.iterrows():
    try: int(row['run']), float(row['duration'])
    except: responses = responses.drop(i)
responses.duration = responses.duration.astype(np.float64)
responses.run = responses.run.astype(int)
df = responses
# df = pd.merge(responses, runs, on='run')
```

```python
q1 = responses['query'] == 'What is the capital of France?'
q1_right = df[q1 & df.response.str.contains('Paris')]
runs.loc[q1_right.run].drop(83)
```

```python
q4 = responses['query'] == 'Who created MongoDB?'
q4_mark = df[q4 & df.response.str.contains('Mark Fontenot').fillna(False)]
runs.loc[q4_mark.run]
```

```python
q2 = responses['query'] == 'What does the C stand for in ACID?'
q2_wrong = df[q2 & ~df.response.str.contains('onsistency').fillna(True)]
runs.loc[q2_wrong.run]
```

```python
q3 = responses['query'] == 'What type of database is Neo4j?'
q3_wrong = df[q3 & ~df.response.str.contains('raph').fillna(True)]
runs.loc[q3_wrong.run]
```
